## imports

```python
import re
import pandas as pd
from pydracor import DraCorAPI
```

## getting lists from DraCor

```python
def get_play_set(corpusname="ger", playname="lessing-emilia-galotti") -> list: 
    play = DraCorAPI().get_play(corpusname, playname) 
    chrctrs = play.characters
    chrcrs_combined_set = set()
    for chrctr in chrctrs:
        chrcrs_combined_set.add(chrctr.name.lower())
        ## add prinz for der prinz and so on
        article_match = re.match(r"(der|die|das) (.+)", chrctr.name.lower())
        if article_match is not None:
            chrcrs_combined_set.add(article_match.group(2).strip())
        chrcrs_combined_set.add(chrctr.id.lower())

    return chrcrs_combined_set
```

```python
get_play_set()
```

```python
get_play_set(playname="schlegel-die-entfuehrte-dose")
```

```python
egal = get_play_set()
```

```python
edose = get_play_set(playname="schlegel-die-entfuehrte-dose")
```
