## getting lists from DraCor

```python
def get_play_list(corpusname="ger", playname="lessing-emilia-galotti") -> list: 
    play = DraCorAPI().get_play("ger","lessing-emilia-galotti") 
    chrctrs = play.characters
    chrcrs_combined_list = []
    for chrctr in chrctrs:
        chrcrs_combined_list.append(chrctr.name)
        chrcrs_combined_list.append(chrctr.id)

    return chrcrs_combined_list
```
